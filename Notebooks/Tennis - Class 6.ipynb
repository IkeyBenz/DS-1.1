```python
import pandas as pd

df = pd.read_csv('../Datasets/tennis.txt', 
                 delimiter="\t", 
                 header=None, 
                 names=['a', 'b', 'c', 'd', 'e'])

df
```

```python
def compare_c_probability(play, wind):
    return len(df[(df.e == play) & (df.d == wind)]) / len(df[df.d == wind])
```

```python
compare_c_probability('Yes', 'Weak')
```
