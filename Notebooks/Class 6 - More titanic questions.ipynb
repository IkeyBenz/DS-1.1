```python
import pandas as pd
import numpy as np
df = pd.DataFrame(pd.read_csv('../Datasets/titanic.csv'))

# Question: What was the probability that the passengers who paid over $100 survived?

total = len(df[df.Fare > 100])
survived = len(df[(df.Fare > 100) & (df.Survived == 1)])

survived / total * 100
```

```python
# Question: What is the probability that a survived passenger was a man?

total = len(df[df.Survived == 1])
men = len(df[(df.Survived == 1) & (df.Sex == 'male')])

men / total * 100
```

```python
# Question: Given a male passenger who was over 30 and didn't survive, what are the odds
#           that they paid less than $25

totalPeople = df[(df.Sex == 'male') & (df.Age > 30) & (df.Survived == 0)]
poor = totalPeople[totalPeople.Fare < 25]

len(poor) / len(totalPeople) * 100
```
